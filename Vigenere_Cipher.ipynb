# **Vigenere Cipher**

Vigenere cypher adalah metode enkripsi string dengan encode Caesar di setiap hurufnya, sesuai dengan huruf kunci.

image.png

```python
# Input huruf

message = input("Input Text: ")
keyword = input("Keyword: ")
encrypt = input("Enkripsi? (Y/N): ")

alphabet = 'abcdefghijklmnopqrstuvwxyz'

if encrypt.lower() == 'y':
  direction = 1

else:
  direction = -1

def vigenere (message, keyword, direction):
  key_index = 0
  encrypted_message = ''

  for char in message:

    if not char.isalpha():
      encrypted_message += ' '

    else:
      index_message = alphabet.find(char)

      # Telusur kata kuncinya
      shift = alphabet.find(keyword[key_index%len(keyword)].lower())

      # Shifting index
      new_index = (index_message + shift*direction)%len(alphabet)

      encrypted_message += alphabet[new_index]

      # Shifting key
      key_index += 1

  print(f'Ciphered Text: {encrypted_message}')

vigenere (message, keyword, direction)
```
